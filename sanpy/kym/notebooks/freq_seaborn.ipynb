## Load a kymograph ROI analysis and plot some stats for each ROI

```python
from sanpy.kym.kymRoiAnalysis import KymRoiAnalysis
```

```python
path = '/Users/cudmore/Desktop/retreat-sept-2024/SSAN Linescan 12.tif'

kra = KymRoiAnalysis(path)

df = kra.getDataFrame()

print(df)
```

```python
import seaborn as sns

sns.set_theme('talk')

sns.set_palette("colorblind")
palette = sns.color_palette()

import matplotlib.pyplot as plt
plt.style.use("dark_background")

fig = plt.figure(figsize=(3, 3), dpi=90)
axs = fig.add_subplot(111)

yStat = 'Peak Inst Freq (Hz)'
# yStat = 'Peak Height'

axScatter = sns.stripplot(data=df,
                                x='ROI Number',
                                y=yStat,
                                hue='ROI Number',
                                alpha=0.6,
                                palette=palette,
                                ax=axs,
                                legend=False,
                                )
axScatter.grid(False)

axScatter = sns.pointplot(data=df,
            x='ROI Number',
            y=yStat,
            hue='ROI Number',
            errorbar=None,  # can be 'se', 'sem', etc
            # errorbar='se',
            linestyle='none',  # do not connect (with line) between categorical x
            marker="_",
            markersize=30,
            # markeredgewidth=3,
            legend=False,
            palette=palette,
            ax=axs
)
axScatter.grid(False)
sns.despine()
```
